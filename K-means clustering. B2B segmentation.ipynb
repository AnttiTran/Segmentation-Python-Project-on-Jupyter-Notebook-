```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

#Using make_blobs tp create sample data
X, y = make_blobs(n_samples=500, n_features=2, centers=3, random_state=23)

# Visualizing the initial data distribution
fig = plt.figure(figsize=(8, 6))
plt.grid(True)
plt.scatter(X[:, 0], X[:, 1], alpha=0.6)
plt.title("Initial Data Distribution")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

# Number of clusters (k)
# Divide the clusters into 3 divisions
k = 3

# Initializing cluster centroids randomly
# Range of [-2, 2] for both dimensions are generated for each cluster by using random initial centroids
clusters = {}
np.random.seed(98)

for idx in range(k):
    center = 2 * (2 * np.random.random((X.shape[1],)) - 1)
    cluster = {
        'center': center,
        'points': []
    }
    clusters[idx] = cluster

# Plotting initial cluster centroids
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], alpha=0.6)
plt.grid(True)
for i in clusters:
    center = clusters[i]['center']
    plt.scatter(center[0], center[1], marker='*', c='red', s=200, label=f"Cluster {i+1} Center")
plt.title("Initial Cluster Centers")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```
